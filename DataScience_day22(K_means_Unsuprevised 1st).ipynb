```python
import pandas as pd
```

```python
df=pd.read_csv("Mall_customers.csv")
df.head()
```

```python
df.columns
```

```python
df=df.rename(columns={
    'Annual Income (k$)':'annual_income',
    'Spending Score (1-100)':'spend_score'
    
})

    
```

```python
df.head()
```

```python
why we have done train test split in the supervised machine laearning?
- xtrain , ytrain
-xtest, ypred,  ytest  for the evaluation we nedd the data to be seperated to the train test 
but in the case of the unsupervised we no need to split the data in the train and test 
```

```python
X=df[['annual_income','spend_score']]
# hamro ma y xaina tei vayrw yo unsupervised 
```

```python
import matplotlib.pyplot as plt
```

```python

plt.scatter(X.annual_income,X.spend_score)
plt.xlabel('Annual_income')
plt.ylabel('Spend_score')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=7) # cluster ko value jati xa teti noii ota cluster banxa rw teo plot ma rw label ma replicate hunccha 
kmeans.fit(X)
```

```python
labels=kmeans.labels_
labels
```

```python
plt.scatter(X.annual_income,X.spend_score, c=labels)
plt.show()
```
