```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("wine.data.csv")
df.head()
```

```python
There are two ways to resuce the dimension of the dataset 
1. Feature Selection=> yeta vayo ko feature haru madnye yauta subset matrw 
select garxa (for example there are 13 columns in this datset , so in feature 
selection it only take 5 6 or any number of the feature using any technique
(yetai vako matrw)
-Curse of dimensionally

2. PCA
- PCA le pani garne ta feature selection noii ho tera in pca yasle chai 
yetai vako column lai select gardaina yesle naya column aafai thapxa 
based on the requirement (jun naya column banxan tini haru lai chai 
princippal component vaninxa rw yesma sabbai column ko info hunccha)
```

```python
df=df.drop('Class',axis=1)
```

```python
df.head()
```

```python
# Scaling 
scaler=StandardScaler()
dfs=scaler.fit_transform(df)
```

```python
dfs=pd.DataFrame(dfs,columns=df.columns)
```

```python
dfs
```

```python
# Applying PCA
pca=PCA()    # if we only need 5 principal component then we use the 
                                    #syntax here :syntax=pca=PCA( n_components=5) 
df_pca=pca.fit_transform(dfs)
```

```python
df_pca
```

```python
df.shape
```

```python
df_pca=pd.DataFrame(df_pca)
```

```python
df_pca.head()
```

```python
aba yesma pani no of component aafno man ley dida teti appropraiate 
hudaina jastai manau hami ley 5 ota principal component matrw chayo 
but teo 5 ota teti enough nahuna sakxa for storing all the information
or may be tini haru ali dherai huna sakxa for storing tesko case ma 
yo have to choose the n_components wisely 
    So to choose the n_components wisely we have two rules :

    1.kaiser Rule
    2.scree plot 
```

```python
yauta column ma yauta eigin value hunccha 
ahilye ko case ma yeta 13 ota column xa so 13 ota eigin value ayunccha
aba pca ley cahi k vanxa vanye jasko eigen value highest xa teslai cahi choose 
garno parxa
```

```python
# to see the eigin values of the each column 
eigen=pca.explained_variance_
eigen
```

# kaiser rule States that consider only those columns whose eigen value is greater than 1 
- aba yo mathi ko output ma heda kati ota column ma value greater than 1 xa teti ota column matrw liney
- yesma only 3 columns have value greater than 1 so take only 3 columns acccording to the kaiser rule 

```python
# Explained Variance
variance=pca.explained_variance_ratio_
variance
```

```python
# evaliation of the above Explained variance 
- here 0.361 says that this column contin 36 % of the information
```

```python
# Scree plot  => yesle chai plot banaune vanxa kay 
=> yesma eigin value ko plot banaune 
```

```python
plt.plot(range(1,14),eigen)
plt.xlabel('no of components')
plt.ylabel('Eigen values')
plt.show()
```
